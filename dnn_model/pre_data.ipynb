```python
from tqdm import *
```

```python
df_trainpool=pd.read_table("/opt/gongxf/python3_pj/Robot/4_cnn_simi/data/train_pool.txt")
```

```python
df_testpool=pd.read_table("/opt/gongxf/python3_pj/Robot/4_cnn_simi/data/test_pool.txt")
```

```python
df_trainpool.head(1)
# df_testpool.head(1)
```

```python
for ii in tqdm(range(len(df_trainpool))):
    df_trainpool['x_pad'][ii]=eval(df_trainpool['x_pad'][ii])
    df_trainpool['pool_x'][ii]=eval(df_trainpool['pool_x'][ii])
```

```python
for ii in tqdm(range(len(df_testpool))):
    df_testpool['x_pad'][ii]=eval(df_testpool['x_pad'][ii])
    df_testpool['pool_x'][ii]=eval(df_testpool['pool_x'][ii])
```

```python
train_xdata=np.array([df_trainpool['pool_x']])[0]
train_xdata=train_xdata.astype(np.float)
train_ydata=df_trainpool['y_pad'].values
test_xdata=np.array([df_testpool['pool_x']])[0]
test_xdata=test_xdata.astype(np.float)
test_ydata=df_testpool['y_pad'].values
```

```python
######################    0—xgboost集成学习方法：   #################
import xgboost as xgb
```

```python
from sklearn import metrics
from xgboost.sklearn import XGBClassifier

clf = XGBClassifier(
silent=0 ,#设置成1则没有运行信息输出，最好是设置为0.是否在运行升级时打印消息。
#nthread=4,# cpu 线程数 默认最大
learning_rate= 0.3, # 如同学习率
min_child_weight=1, 
# 这个参数默认是 1，是每个叶子里面 h 的和至少是多少，对正负样本不均衡时的 0-1 分类而言
#，假设 h 在 0.01 附近，min_child_weight 为 1 意味着叶子节点中最少需要包含 100 个样本。
#这个参数非常影响结果，控制叶子节点中二阶导的和的最小值，该参数值越小，越容易 overfitting。
max_depth=3, # 构建树的深度，越大越容易过拟合
gamma=0,  # 树的叶子节点上作进一步分区所需的最小损失减少,越大越保守，一般0.1、0.2这样子。
subsample=1, # 随机采样训练样本 训练实例的子采样比
max_delta_step=0,#最大增量步长，我们允许每个树的权重估计。
colsample_bytree=1, # 生成树时进行的列采样 
reg_lambda=1,  # 控制模型复杂度的权重值的L2正则化项参数，参数越大，模型越不容易过拟合。
#reg_alpha=0, # L1 正则项参数
#scale_pos_weight=1, #如果取值大于0的话，在类别样本不平衡的情况下有助于快速收敛。平衡正负权重
objective= 'multi:softmax', #多分类的问题 指定学习任务和相应的学习目标
#num_class=10, # 类别数，多分类与 multisoftmax 并用
n_estimators=111, #树的个数
seed=10 #随机种子
#eval_metric= 'auc'
)
clf.fit(train_xdata,train_ydata,eval_metric='auc')
y_true, y_pred = test_ydata, clf.predict(test_xdata)
print("Accuracy : %.4g" % metrics.accuracy_score(y_true, y_pred))
#回归
#m_regress = xgb.XGBRegressor(n_estimators=1000,seed=0)
```

```python
######################    1—bagging集成学习方法：随机森林   #################
robot_rf=RandomForestClassifier(n_estimators=111,criterion='gini',max_features='auto',min_samples_split=2,min_samples_leaf=1,
                                min_weight_fraction_leaf=0.0,n_jobs=3,random_state=10
                                )
robot_rf.fit(train_xdata,train_ydata)
test_predict=robot_rf.predict(test_xdata)
Accuracy=sum(test_predict==test_ydata)/len(test_ydata)*100
Accuracy
# a=robot_rf.predict_proba(test_xdata)  
# np.argmax(a[0])
# max(a[0]) 
```

```python
################       2-sklearn：多层感知器  #########
from sklearn.linear_model import Perceptron
```

```python
robot_ppn = Perceptron(penalty='l1',max_iter=20,n_jobs=10)  # y=w.x+b
robot_ppn.fit(train_xdata, train_ydata)
test_predict=robot_ppn.predict(test_xdata)
Accuracy=sum(test_predict==test_ydata)/len(test_ydata)*100
print("Accuracy",Accuracy)
```

```python
################      3-sklearn：多层感知器+网格搜索############
from sklearn.model_selection import GridSearchCV
#把要调整的参数以及其候选值 列出来；
param_grid = {"max_iter":[1,5],
             "penalty":['l2','l1','elasticnet']}
print("Parameters:{}".format(param_grid))
```

```python
grid_search = GridSearchCV(Perceptron(n_jobs=10),param_grid,cv=5) #实例化一个GridSearchCV类
# X_train,X_test,y_train,y_test = train_test_split(iris.data,iris.target,random_state=10)
grid_search.fit(train_xdata,train_ydata) #训练，找到最优的参数，同时使用最优的参数实例化一个新的SVC estimator。
print("Test set score:{:.2f}".format(grid_search.score(test_xdata,test_ydata)))
print("Best parameters:{}".format(grid_search.best_params_))
print("Best score on train set:{:.2f}".format(grid_search.best_score_))
```

```python
##########       4-sklearn-boosting:GBDT       #######
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import cross_validation, metrics
from sklearn.grid_search import GridSearchCV
```

```python
robot_gbm = GradientBoostingClassifier(n_estimators=11,random_state=10)
robot_gbm.fit(train_xdata,train_ydata)
test_predict = robot_gbm.predict(test_xdata)
Accuracy=sum(test_predict==test_ydata)/len(test_ydata)*100
Accuracy
```

```python
#############      5-sklearn:adaboost-CART树集成       ##################
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import zero_one_loss
from sklearn.ensemble import AdaBoostClassifier
```

```python
n_estimators = 111      #弱学习器的个数
learning_rate = 1.0

# 决策树桩
dt_stump = DecisionTreeClassifier(min_samples_leaf=1)
dt_stump.fit(train_xdata, train_ydata)
##决策树桩的生成集成学习器
robot_ada = AdaBoostClassifier(base_estimator=dt_stump, learning_rate=learning_rate,
                                      n_estimators=n_estimators, algorithm='SAMME')
robot_ada.fit(train_xdata, train_ydata)
test_predict=robot_ada.predict(test_xdata)
Accuracy = sum(test_predict == test_ydata) / len(test_ydata) * 100
print("---Accuracy---", Accuracy)
```
