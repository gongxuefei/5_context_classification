```python
#分层抽样生成训练样本与测试样本
train_stratified_sampling = pd.DataFrame(columns=['class', 'question'])
test_stratified_sampling = pd.DataFrame(columns=['class', 'question'])
for class_text in class_count.index:
    df_tmp=dff[dff['class']==class_text]
    train_tmp, test_tmp = train_test_split(df_tmp, test_size = 0.3)
#     print(len(train_tmp),len(test_tmp))
    train_stratified_sampling = train_stratified_sampling.append(train_tmp)
    test_stratified_sampling = test_stratified_sampling.append(test_tmp)
#     print(len(train_stratified_sampling),len(test_stratified_sampling))
#保存分层抽样的样本
train_stratified_sampling.to_csv('./data/train_stratified_sampling.txt',sep='\t',index=False)
test_stratified_sampling.to_csv('./data/test_stratified_sampling.txt',sep='\t',index=False)
```

```python
df_train=pd.read_table("./data/train_keyword.txt", names=['class','question'])
```

```python
df_train.head(1)
```

```python
# df_train.to_csv('./data/train.txt',sep='\t',index=False)
```

```python
train_dict={}
for i in range(len(df_train)):
    if df_train['class'][i] in train_dict.keys():
        train_dict[df_train['class'][i]].append(df_train['question'][i])
    else:
        train_dict[df_train['class'][i]]=[df_train['question'][i]]
train_dict
```

```python
train_count=df_train.groupby('class').count()
```

```python
train_count
```

```python
#对训练数据进行上采样
import random
num_sample=2000
ff=open("./data/train_keyword.txt",mode='a+',encoding='utf-8')
for ii in range(len(train_count)):
    if train_count['question'][ii]<num_sample:
#         print(train_count['question'][ii])
        num=len(train_dict[train_count.index[ii]])
#         print("num",num)
        for i in range(num_sample-train_count['question'][ii]):
#             print("i",i)
#             print(train_count.index[ii]+'\t'+train_dict[train_count.index[ii]][random.randint(0,num-1)])
            ff.write(train_count.index[ii]+'\t'+train_dict[train_count.index[ii]][random.randint(0,num-1)]+'\n')
ff.close()
```
